## Scikit-learn의 주요 특징
- 다양한 알고리즘
    - 분류, 회귀, 클러스터링, 차원 축소 등 머신러닝의 주요 알고리즘을 모두 포함
- 통일된 인터페이스
    - 모든 알고리즘이 동일한 방식으로 사용될 수 있도록 설계되어 있어 학습, 예측, 평가 과정이 매우 직관적
- 효율성
    - NumPy, SciPy와 같은 선형 대수 라이브러리와 최적화된 알고리즘을 기반으로 하여 빠르고 효율적인 성능을 제공
- 확장성
    - 다른 파이썬 데이터 과학 라이브러리와 쉽게 통합
- 활발한 커뮤니티
    - 많은 개발자들이 참여하고 있으며, 지속적으로 새로운 기능이 추가

## Scikit-learn 주요 모듈
- sklearn.datasets
    - 예제 데이터 세트
- sklearn.metrics
    - 성능평가
- sklearn.preprocessing
    - 데이터 전처리(범주형 인코딩, 스케일링 등)
- sklearn.feature_selection
    - 특성 선택
- sklearn.model_selection
    - 데이터 분리, 하이퍼 파라미터 튜닝
- sklearn.ensemble
    - 앙상블 알고리즘
- sklearn.cluster
    - 군집분석 관련 알고리즘
- sklearn.decomposition
    - 차원축소 관련 알고리즘



## Scikit-learn 모델
- 주어진 문제를 해결하기 위한 다양한 모델들이 있음
- Scikit-learn의 지도학습 관련 모델의 대부분은 회귀와 분류 문제에 모두 사용 가능함
- `Classifier`: 분류 알고리즘을 구현한 모델 클래스
    - ex) `DecisionTreeClassifier`
- `Regressor`: 회귀 알고리즘을 구현한 모델 클래스
    - ex) `DecisionTreeRegressor`



## Scikit-learn 사용 방법
- 머신러닝 단계별 적절한 모듈을 선택하여 클래스, 함수를 Import 하여 사용
- scikit-learn은 클래스 사용법을 통일된 방식으로 제공
- 모델 학습을 위한 메서드
    - `fit`: 모델 학습
    - `predict`: 모델 예측
    - `predict_proba`: 모델이 예측한 확률
- 데이터 변환을 위한 메서드
    - `fit`: 데이터를 변환하기 위한 사전작업
    - `transform`: 데이터를 변환
    - `fit_transform`: fit 과 transform 한번에 수행

# 특성 공학(Feature Engineering)
- 학습데이터에 대해 Feature Engineering 과정에서 수행했던 작업을 평가데이터에  동일한 방식으로 수행할 것

## Feature Extraction
- 특성으로 사용할 변수 추출하기

````python
train.columns
````

````python
cols=['pclass','gender','age','sibsp','parch','fare','embarked']
train_ft=train[cols].copy()
test_ft=test[cols].copy()
````

````python
train_ft.shape, test_ft.shape
````

````python
train_ft
````

````python
train_ft['가족수'] = train_ft['sibsp'] + train_ft['parch']
test_ft['가족수'] = test_ft['sibsp'] + test_ft['parch']
````

````python
train_ft
````

## Data Cleaning
- 결측치 및 이상치 처리와 같은 데이터 정제

````python
train.isnull().sum()
````

````python
train_ft.isnull().sum()
````

````python
test_ft.isnull().sum()
````

#### 학습데이터에서 얻은 통계량으로 결측치를 처리해야 한다.

````python
age_mean = train_ft['age'].mean()
fare_median = train_ft['fare'].median()
embarked_mode = train_ft['embarked'].mode()[0]
````

#### 학습데이터 결측치 처리

````python
train_ft['age']=train_ft['age'].fillna(age_mean)
````

#### 테스트 데이터 결측치 처리

````python
test_ft['age']=test_ft['age'].fillna(age_mean)
test_ft['fare']=test_ft['fare'].fillna(fare_median)
test_ft['embarked']=test_ft['embarked'].fillna(embarked_mode)
````

````python
train_ft.isnull().sum().sum(  ), test_ft.isnull().sum().sum()
````

````python
train_ft
````

## Feature Encoding

- 범주형 변수 원핫인코딩하여 특성으로 추가하기
    - 원핫인코딩(One-Hot Encoding)이란?
        - 범주를 많은 0과 하나의 1값으로 범주를 수치화하는 인코딩 기법

20221213154823.png

````python
from sklearn.preprocessing import OneHotEncoder
````

````python
cols = ['gender','embarked']
enc = OneHotEncoder()
train[cols]
````

````python
enc.fit(train[cols])
````

````python
enc.transform(train_ft[cols]).toarray()
````

````python
enc.get_feature_names_out() # 인코딩 데이터 컬럼명
````

````python
tmp = pd.DataFrame(
    enc.transform(train_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
train_ft = pd.concat([train_ft, tmp], axis=1).drop(columns=cols)
train_ft.head()
````

````python
tmp = pd.DataFrame(
    enc.transform(test_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
test_ft = pd.concat([test_ft, tmp], axis=1).drop(columns=cols)
test_ft.head()
````

## Feature Scaling
- Min-Max Scaling
    - 데이터 수치의 범위를 바꿔주는 정규화 스케일링 기법입니다.
    - 데이터 수치의 범위를 0 ~ 1 사이로 바꿔준다.
$$
\frac{x-Min(X)}{Max(X)-Min(X)}
$$

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
````

````python
train_ft[train_ft.columns] = scaler.fit_transform(train_ft)
train_ft.head()
````

````python
test_ft[test_ft.columns] = scaler.transform(test_ft)
test_ft.head()
````

````python
target = train["survived"]
target
````

# 데이터 분리(Data Split)
- 학습데이터와 검증데이터 분리

````python
from sklearn.model_selection import train_test_split
## 학습용 입력데이터, 검증용 입력데이터, 학습용ㅇ 정답데이터, 검즈용ㅇ 정답데이터

x_train, x_valid, y_train, y_valid = train_test_split(train_ft, target, test_size=0.2, random_state=42)

x_train.shape, x_valid.shape, y_train.shape, y_valid.shape

#### 오류나는건 앞시간에 못들은 target변수가 있다는 것
````

# 모델 학습(Train Model)


````python
from sklearn.linear_model import LogisticRegression
````

````python
model = LogisticRegression(random_state=42)
````

````python
model.fit(x_train, y_train)
````

````python
model.predict(x_valid)
````

````python
model.predict_proba(x_valid)[:,1]
````

#### 사전에 검증평가 해보기

````python
pred = model.predict(x_valid)
pred
````

````python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid, pred)
````

# 모델 평가(Model Evaluation)

````python

## 테스트 데이터에 대해서 평가하기 전에 전체 데이터에 대해서 다시 학습
model.fit(train_ft, target)
````

````python
pred = model.predict(test_ft)
pred
````

````python
pd.DataFrame(pred, columns=['target']).to_csv('jerry_logisticmodel.csv', index=False)
````
